### Reglas de aprendizaje y representaciones de los datos

Para definir el Deep Learning y entender la diferencia entre el acercamiento entre éste y el ML, primero debemos tener una idea intuitiva de qué es lo que hacen los algoritmos de ML.

Anteriormente dijimos que el ML descubre reglas para ejecutar una tarea de procesamiento de datos, dado ejemplos de lo que debe esperar. Entonces, para hacer ML, necesitamos de tres cosas:

1. Datos de entrada (input)

2. Ejemplos de la salida esperada (outputs)

3. Una forma de medir si el algoritmo está haciendo un buen trabajo

Este último requisito es necesario para determinar la distancia entre el output actual del algoritmo y su output esperado. La medida se utiliza como una señal de retroalimentación para ajustar la forma en la que el algoritmo funciona. Este paso de ajuste es a lo que le llamamos *aprendizaje*.

Un modelo de ML transforma sus datos de entrada en salidas con significado, un proceso que es "aprendido" de la exposición a ejemplos conocidos de inputs y outputs. Por lo tanto, el problema central en el ML y en DL, es el de *transformar los datos significativamente*; es decir, aprender *representaciones* útiles de los datos de entrada.

Pero, ¿qué es una representación? Simplemente es una forma distinta de ver a los datos. Por ejemplo, una imagen a color se puede codificar en formato RGB o en el formato HSV, las cuales son dos representaciones distintas de los mismos datos.

Algunas tareas pueden ser más sencillas en una representación que en la otra, por lo que en gran parte, los modelos de ML tratan de encontrar las representaciones adecuadas para sus datos de entrada: transformaciones de los datos que los hagan más manejables para la tarea a realizar.


*Aprender* en el contexto de ML, describe un proceso de búsqueda automático para transformaciones de datos que producen una representación útil de los mismos, guiado por alguna señal de retroalimentación. 

### El Deep en Deep Learning

Deep Learning es un subcampo específico del ML. El "deep" en el nombre no es una referencia a algún tipo de entendimiento más profundo que se alcance con este enfoque; sino que representa la idea de capas sucesivas de representaciones. Cuántas capas contribuyen al modelo de los datos es lo que se conoce como la *profundidad* del modelo. 

El DL moderno involucra a menudo decenes o incluso cientas de capas sucesivas de representación, y todas aprenden de forma automática a partir de la exposición a datos de entrenamiento. En el DL, estas representaciones a capas son aprendidas a través de modelos llamados *redes neuronales*. 

El término 'red neuronal' proviene de la neurobiología, y si bien algunos de los conceptos centrales en el DL fueron desarrollados inspirados en nuestro entendimiento del cerebro (en específico de la corteza visual), los modelos de DL no son modelos del cerebro. Para nuestros propósitos, el DL es un marco de trabajo matemático para aprender representaciones de los datos.

### Redes Neuronales Artificiales

Las Redes Neuronales (RN), también conocidas como Redes Neuronales Artificiales, son un subconjunto del ML y son el corazón de los algoritmos de DL. 

Las RN están compuestas de capas de nodos, con una capa de entrada, una o más capas ocultas, y una capa de salida.

nn.png

Cada nodo, también llamado *neurona*, está conectado con otro y tiene dos parámetros asociados, un *peso* $w$, y un *bias* o *umbral* $b$.

Si la salida de cualquier neurona individual es mayor que el $b$ especificado, la neurona se *activa*, y envía los datos que contiene a la siguiente capa de la red.

En caso contrario, los datos no pasan de una capa a la otra.

Poniéndolo en un contexto matemático, podemos pensar en cada neurona individual como su propio modelo de regresión lineal, compuesto por datos de entrada $x_i$, $w_i$, $b$ y una salida. La expresión sería algo así:

$$ \sum_{i=1}^n w_i x_i + b = w_1 x_1 + w_2 x _2 + \dots + w_n x_m + b $$

Una vez que la capa de entrada se determina, los pesos de cada neurona son asignados. Estos pesos ayudan a determinar la importancia de cualquier variable dada, mientras mayor sea el valor del peso de una variable, mayor es la contribución a la salida en comparación al resto de las variables.

Podemos ver que todas las entradas son multiplicadas por sus respectivos pesos y después se suman. Después, el resultado de esta suma es ingresado a una *función de activación*, que determina la salida. 

Si la salida es mayor que cierto *umbral*, entonces la neurona es activada, pasando los datos a la siguiente capa de la red.

$$ \text{salida} = f(x) = \begin{cases}
1 & \text{si} & \sum w_i x_i + b \geq 0\\
0 & \text{si} & \sum w_i x_i + b < 0 \\
\end{cases} $$

Por lo que la salida de una neurona se convierte en la entrada de la siguiente neurona en la siguiente capa. Este proceso de pasar información de una capa a la siguiente define a la red neuronal como una red de *propagación hacia adelante* (*feedforward*).

Poniendolo en un ejemplo muy sencillo y concreto, utilizando valores binarios.

¿Deberíamos ir a una fiesta? (Sí: 1, No: 0)
La decisión de ir o no es nuestra predicción, a menudo denotada como $\hat{y}$. 

Supongamos que hay tres factores que determinan nuestra decisión:

1. Si vamos con amigos (Sí: 1, No: 0)
2. Si habrá comida (Sí: 1, No: 0)
3. Si tenemos transporte (Sí: 1, No: 0)

Supongamos que tenemos las siguientes entradas:

- $x_1 = 1$, si vamos con amigos
- $x_2 = 0$, no habrá comida (buuu)
- $x_3 = 1$, sí nos van a dar un aventón

Ahora, debemos asignar algunos pesos para determinar la importancia. Somos sociables, pero somos más comelones, y nos es irrelevante si nos dormimos en la fiesta o en nuestra casa.

- $w_1 = 4$
- $w_2 = 5$
- $w_2 = 2$

Y por último, supongamos que, en una escala del 1 al 5, podemos ir o no a la fiesta, no hay problema, así que nuestro valor umbral lo ponemos en $3$. Con esa información, tenemos:



```python
import numpy as np 

x_i = np.array([1,0,1])

w_i = np.array([4,5,2])

b = 3
a = np.dot(x_i, w_i) + b

print('a =', a)

if a >= 3:
  print('f(x) = 1, si vamos')
else:
  print('f(x) = 0, nos quedamos')
```

El resultado es que sí vamos a la fiesta, sin embargo, si ajustamos los pesos $w$ o el valor umbral $b$, podríamos obtener otra respuesta. 

En este ejemplo se utilizaron los *perceptrones*, el tipo más sencillo de red neuronal y que se ilustra en la siguiente imagen.

Sin embargo, este esquema se puede extender a otro tipo de tareas como el reconocimiento de imágenes o la clasificación, que utilizan *aprendizaje supervisado*, o conjuntos de datos etiquetados, para entrenar al algoritmo.

Como ya mencionamos anteriormente, conforme entrenamos el modelo, necesitamos evaluar su desempeño, y lo hacemos a través de una *función de costo*. Una de las funciones de costo más comunes es el *error cuadrático medio* (*MSE*), dado por:

$$ \text{Funcion de costo} = \text{MSE} = \frac{1}{2m} \sum_{i =1}^m \left(  \hat{y}_i - y_i \right)^2$$

donde $\hat{y}$ es el resultado predecido, $y$ el valor real, y $m$ el número de muestras.

El objetivo consiste en minimizar la función de costo para asegurar el correcto ajuste para cualquier observación dada. Conforme el modelo ajusta los $w$ y los $b$, utiliza la función de costo y el aprendizaje por refuerzo para alcanzar un mínimo local. 

El proceso en el cual el algoritmo ajusta sus pesos es a través del *gradiente descendiente*, que permite al modelo determinar la dirección a tomar para minimizar la función de costo. Gradualmente, los parámetros del modelo se ajustan para converger al mínimo global.

## Recapitulando

Hasta este momento, sabemos que en escencia, el ML mapea datos de entrada (como imágenes) a ciertas salidas objetivo (como una etiqueta que diga "gato"), y que lo hace al observar muchos ejemplos de entradas y salidas.

También, vimos que las redes neuronales profundas lo hacen a través de una secuencia profunda de transformaciones de datos (capas) y que esas transformaciones son aprendidas al exponerla a muchos ejemplos.

Concretamente, el "qué" le hace una capa a los datos de entrada está almacenado en la capa de pesos, que en realidad son un montón de números. 

Formalmente, decimos que la transformación aplicada por una capa está *parametrizada* por sus pesos. 

En este contexto, *aprender* significa encontrar un conjunto de valores para los pesos de todas las capas en una red, tales que la red aprenda a mapear correctamente las entradas a sus salidas respectivas.

imagen.png



¿Cómo encontrar los parámetros adecuados? Midiendo qué tan lejos se encuentran las salidas obtenidas de las salidas esperadas. Eso es trabajo de la *función de costo de la red*. La función de costo toma las prediciones de la red y las salidas reales y calcula la distancia entre ellos, esa métrica se conoce como *loss score*.

imagen.png

El truco fundamental en el DL consiste en utilizar ese valor de la función de costo como una señal de retroalimentación para ajustar un poquito el valor de los pesos, en una dirección que logre bajar el *loss score* para el ejemplo actual. Este ajuste es el trabajo del *optimizador*, que implementa el famoso algoritmo *backpropagation*. Veremos más adelante cómo funciona.

imagen.png

Inicialmente, los pesos de la red son asignados aleatoriamente, para que la red implemente una serie de transformaciones aleatorias. Claro, que su salida puede estar muy lejos de lo que debería, y el loss score va a ser muy alto. Sin embargo, con cada ejemplo que la red procesa, los pesos se van ajustando poco a poco hacia la dirección correcta, lo que disminuye el loss score.

Eso constituye el *ciclo de entrenamiento*, el cual, repetido el número de veces suficientes (descenas de iteraciones sobre miles de ejemplos), conlleva a valores de los pesos que minimizan la función de costo. 

Una red que tiene un costo mínimo es aquella cuyas salidas son lo más cercano posible a los objetivos.

Eso es una red neuronal entrenada.

## Un primer vistazo a una red neuronal

Veamos un ejemplo en concreto de una red neuronal que utiliza la biblioteca de Python, Keras, para aprender a clasificar dígitos escritos a mano.

El problema que intentaremos resolver es clasificar imágenes en escala de grises de dígitos escritos a mano (de 28x28 pixeles) en sus 10 posibles categorías (0 a 9). Se utilizará el conjunto MNIST, un clásico de la comunidad de ML. 

Consiste en 60,000 imágenes de entrenamiento, más 10,000 imágenes de prueba. 

![MNIST Dataset](https://www.researchgate.net/profile/Steven-Young-5/publication/306056875/figure/fig1/AS:393921575309346@1470929630835/Example-images-from-the-MNIST-dataset.png)

En ML, una *categoría*, en un problema de clasificación es llamada una *clase*. Los datos se llaman *muestras* y la clase asociada a una muestra específica es conocida como *etiqueta*.

Bien, el conjunto MNIST viene precargado en Keras, por lo que no es necesario descargarlo manualmente.
Viene pre procesado como un conjunto de cuatro arrays de NumPy.

```python
#pip install tensorflow
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

Las variables `train_images` y `train_labels` conforman el conjunto de entrenamiento, es decir, los datos de los que el modelo va a aprender. 

El modelo después va a ser probado en el conjunto de prueba, que corresponde a las variables `test_images` y `test_labels`.

En este caso, las imágenes están codificadas como arrays de NumPy, y las etiquetas son un array de dígitos, que van de 0 a 9. Las imágenes y las etiquetas tienen una correspondencia de uno a uno.

¿Qué tenemos que hacer? Primero, vamos a alimentar a la red neuronal con los datos de entrenamiento. La red entonces va a aprender a asociar imágenes y etiquetas. Finalmente, le pediremos a la red que haga predicciones para el conjunto de prueba, y verificaremos que las predicciones coincidan con las etiquetas de prueba.

Construyamos la red

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(42, activation = 'relu'),
    # layers.Dense(420, activation = 'linear'),
    # layers.Dense(69, activation = 'relu'),
    layers.Dense(10, activation = 'softmax')
])
```

La pieza fundamental de las redes neuronales es la *capa* (*layer*). Podemos pensar en una capa como en un filtro para los datos: algunos pasan y salen de forma más útil. 

Específicamente, las capas extraen *representaciones* de los datos con los que se alimentan.

Aquí, el modelo consiste en una secuencia de dos capas `Dense` que como su nombre lo indica, son capas de neuronas que están densamente conectadas (también llamadas *completamente conectadas*). La segunda capa es una capa de *clasificación softmax* de 10 salidas, que significa que regresará un array de 10 probabilidades (que suman 1). Cada entrada será la probabilidad de que la imagen actual pertenezca a una de las 10 clases de dígitos.

Para hacer que el modelo esté listo para entrenar, necesitamos de tres elementos más en el paso de *compilación*:

- Un *optimizador*. El mecanismo a través del cual el modelo va a actualizarse basado en los datos de entrenamiento que se le presentan, así como para mejorar su desempeño.

- Una *función de costo*. Cómo es que el modelo va a medir su desempeño en los datos de entrenamiento, y por lo tanto cómo va a poder dirigirse en la dirección correcta.

- *Métricas* para monitorear durante el entrenamiento y el test. En esta ocasión, sólo nos va a importar la *precisión* (*accuracy*), es decir, la fracción de imágenes correctamente clasificadas.

```python
model.compile(optimizer = 'adam',
                loss = 'sparse_categorical_crossentropy',
                metrics = ['accuracy'])
```

Antes de entrenar, vamos a procesar un poquito más los datos al cambiarles la forma que el modelo espera recibir y escalando de tal forma que todos los valores estén en el intervalo [0,1].

Anteriormente vimos que las imágenes de entrenamiento están almacenadas en un array de tamaño `(60000,28,28)` de tipo `uint8` con valores en el intervalo [0,255].

Vamos a transformarlos en un array con valores `float32` de tamaño `(60000,28,28)` con valores entre 0 y 1.

```python
test_digits = test_images[0:10]
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(test_digits[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255.0

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255.0
```

Ahora sí, podemos entrenar el modelo, en Keras lo hacemos llamando al método `fit()` del modelo, porque *ajustamos* el modelo a sus datos de entrenamiento.

```python
model.fit(train_images, train_labels, epochs = 3, batch_size = 250)
```

Notemos qué nos arroja ese comando. Tenemos dos cantidades mostradas en pantalla durante el entrenamiento: el loss del modelo y su accuracy sobre los datos de entrenamiento. Notamos que al final alcanzamos una precisión del 0.9076 ($\approx$ 91%) de los datos de entrenamiento.

Una vez que se entrenó el modelo, podemos usarlo para predecir las probabilidades de a qué clase pertencen *nuevos* dígitos. Es decir, vamos a tomar imágenes del conjunto de prueba.

```python
test_digits = test_images[0:10]

test_digits
```

```python
predictions = model.predict(test_digits)
```

```python
predictions[0]
```

Aquí si ocurre la magia. Cada posición de índice `i` en el resultado anterior corresponde a la probabilidad de que el dígito de la imagen `test_digits[0]` pertenezca a la clase `i`.

Es decir, que el 1 se encuentra en la posición con índice i = 7, así que de acuerdo a nuestro modelo, la imagen de prueba debe corresponder a un 7 :D.


```python
predictions[0].argmax()
```

Y ahora sí, revisamos que la etiqueta coincida:

```python
test_labels[0]
```

```python
predictions[5]
```

```python
predictions[5].argmax()
```

```python
test_labels[5]
```

¡SÍ ES UN 7! Pero pudo haber sido una coincidencia, ¿no? En promedio, ¿qué tan bueno es el modelo clasificando dígitos que nunca antes ha visto? Podemos verlo calculando el promedio de la exactitud sobre todo el conjunto de entrenamiento.

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)

print(f'Exactitud del test: {round(test_acc,3)}')
```

¡Ups! La exactitud del conjunto de prueba es de alrededor del 82.3%, mucho más baja que la exactitud del conjunto de entrenamiento.

Esta brecha entre exactitudes es un ejemplo de *sobreajuste* (*overfitting*): el hecho de que los modelos de machine learning tienden a desempeñarse peor sobre nuevos datos que sobre el conjunto de entrenamiento. Más sobre eso después.
